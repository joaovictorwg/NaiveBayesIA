### Imports

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import math
from sklearn.preprocessing import StandardScaler
```

### Load datasets

```python

# Carrega a base de dados Iris
iris = datasets.load_iris()

# Converte os dados para um DataFrame do Pandas
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['class'] = iris.target
iris_df['class'] = iris_df['class'].map(dict(enumerate(iris.target_names)))

# Exibe o DataFrame completo
print(iris_df.to_string(index=True))
data, target = iris.data, iris.target

# Filtrar Setosa
setosa_indices = np.where(target == 0)[0]
setosa_data = data[setosa_indices]
setosa_target = target[setosa_indices]

# Dividir Setosa em treino e teste
setosa_data_train, setosa_data_test, setosa_target_train, setosa_target_test = train_test_split(setosa_data, setosa_target, test_size=0.2, random_state=42)

# Filtrar Versicolor
versicolor_indices = np.where(target == 1)[0]
versicolor_data = data[versicolor_indices]
versicolor_target = target[versicolor_indices]

# Dividir Versicolor em treino e teste
versicolor_data_train, versicolor_data_test, versicolor_target_train, versicolor_target_test = train_test_split(versicolor_data, versicolor_target, test_size=0.2, random_state=42)

# Filtrar Virginica
virginica_indices = np.where(target == 2)[0]
virginica_data = data[virginica_indices]
virginica_target = target[virginica_indices]

# Dividir Virginica em treino e teste
virginica_data_train, virginica_data_test, virginica_target_train, virginica_target_test = train_test_split(virginica_data, virginica_target, test_size=0.2, random_state=42)

# Inicialização do scaler
scaler = StandardScaler()

# Normalização para Setosa
setosa_data_train_norm = scaler.fit_transform(setosa_data_train)
setosa_data_test_norm = scaler.transform(setosa_data_test)

# Normalização para Versicolor
versicolor_data_train_norm = scaler.fit_transform(versicolor_data_train)
versicolor_data_test_norm = scaler.transform(versicolor_data_test)

# Normalização para Virginica
virginica_data_train_norm = scaler.fit_transform(virginica_data_train)
virginica_data_test_norm = scaler.transform(virginica_data_test)
```

### Calcular Média dos Atributos
#### Fórmula da Média
$$
Média = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

```python
#   MEDIA SEPAL LENGTH
def calcular_media_atributos(data):
    # MEDIA SEPAL LENGTH
    sepal_length_data = data[:,0]
    total_sepal_length = 0

    for length in sepal_length_data:
        total_sepal_length += length
    media_sepal_length = total_sepal_length / len(sepal_length_data)

    # MEDIA SEPAL WIDTH
    sepal_width_data = data[:,1]
    total_sepal_width = 0

    for width in sepal_width_data:
        total_sepal_width += width
    media_sepal_width = total_sepal_width / len(sepal_width_data)

    #   MEDIA PETAL LENGTH
    petal_length_data = data[:,2]
    total_petal_length = 0

    for length in petal_length_data:
        total_petal_length += length
    media_petal_length = total_petal_length / len(petal_length_data)

    #   MEDIA PETAL WIDTH
    petal_width_data = data[:,3]
    total_petal_width = 0

    for width in petal_width_data:
        total_petal_width += width
    media_petal_width = total_petal_width / len(petal_width_data)

    return media_petal_length, media_petal_width, media_sepal_length, media_sepal_width

# Calcular e imprimir médias dos atributos para cada classe

# Setosa
media_petal_length_setosa, media_petal_width_setosa, media_sepal_length_setosa, media_sepal_width_setosa = calcular_media_atributos(setosa_data_train_norm)
print(f"Média dos atributos para Setosa:")
print(f"Petal Length: {media_petal_length_setosa}")
print(f"Petal Width : {media_petal_width_setosa}")
print(f"Sepal Length: {media_sepal_length_setosa}")
print(f"Sepal Width : {media_sepal_width_setosa}")
print()

# Versicolor
media_petal_length_versicolor, media_petal_width_versicolor, media_sepal_length_versicolor, media_sepal_width_versicolor = calcular_media_atributos(versicolor_data_train_norm)
print(f"Média dos atributos para Versicolor:")
print(f"Petal Length: {media_petal_length_versicolor}")
print(f"Petal Width : {media_petal_width_versicolor}")
print(f"Sepal Length: {media_sepal_length_versicolor}")
print(f"Sepal Width : {media_sepal_width_versicolor}")
print()

# Virginica
media_petal_length_virginica, media_petal_width_virginica, media_sepal_length_virginica, media_sepal_width_virginica = calcular_media_atributos(virginica_data_train_norm)
print(f"Média dos atributos para Virginica:")
print(f"Petal Length: {media_petal_length_virginica}")
print(f"Petal Width : {media_petal_width_virginica}")
print(f"Sepal Length: {media_sepal_length_virginica}")
print(f"Sepal Width : {media_sepal_width_virginica}")
```

## Calcular Variância e Desvio Padrão

#### Fórmula da Variância
$$
Variância = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

#### Fórmula do Desvio Padrão
$$
Desvio\ Padrão = \sqrt{Variância}
$$

```python
#   VARIANCIA SEPAL LENGTH
def calcular_variancia_desvio(data, media_sepal_length, media_sepal_width, media_petal_length, media_petal_width):

    sepal_length_data = data[:,0]
    total_sepal_length = 0

    for length in sepal_length_data:
        total_sepal_length += (length - media_sepal_length) ** 2
    variancia_sepal_length = total_sepal_length / len(sepal_length_data)

    # DESVIO PADRÃO SEPAL LENGTH
    desvio_padrao_sepal_length = math.sqrt(variancia_sepal_length)

    # VARIANCIA SEPAL WIDTH
    sepal_width_data = iris.data[:,1]
    total_sepal_width = 0

    for width in sepal_width_data:
        total_sepal_width += (width - media_sepal_width) ** 2
    variancia_sepal_width = total_sepal_width / len(sepal_width_data)

    # DESVIO PADRÃO SEPAL WIDTH
    desvio_padrao_sepal_width = math.sqrt(variancia_sepal_width)

    #   VARIANCIA PETAL LENGTH
    petal_length_data = iris.data[:,2]
    total_petal_length = 0

    for length in petal_length_data:
        total_petal_length += (length - media_petal_length) ** 2
    variancia_petal_length = total_petal_length / len(petal_length_data)

    # DESVIO PADRÃO PETAL LENGTH
    desvio_padrao_petal_length = math.sqrt(variancia_petal_length)

    #   VARIANCIA PETAL WIDTH
    petal_width_data = iris.data[:,3]
    total_petal_width = 0

    for width in petal_width_data:
        total_petal_width += (width - media_petal_width) ** 2
    variancia_petal_width = total_petal_width / len(petal_width_data)

    # DESVIO PADRÃO PETAL WIDTH
    desvio_padrao_petal_width = math.sqrt(variancia_petal_width)

    

    return (variancia_sepal_length, desvio_padrao_sepal_length,
            variancia_sepal_width, desvio_padrao_sepal_width,
            variancia_petal_length, desvio_padrao_petal_length,
            variancia_petal_width, desvio_padrao_petal_width)

# SETOSA
(variancia_sepal_length_setosa, desvio_padrao_sepal_length_setosa,
 variancia_sepal_width_setosa, desvio_padrao_sepal_width_setosa,
 variancia_petal_length_setosa, desvio_padrao_petal_length_setosa,
 variancia_petal_width_setosa, desvio_padrao_petal_width_setosa) = calcular_variancia_desvio(setosa_data_train_norm, media_petal_length_setosa, media_petal_width_setosa, media_sepal_length_setosa, media_sepal_width_setosa)
# Para Versicolor
(variancia_sepal_length_versicolor, desvio_padrao_sepal_length_versicolor,
 variancia_sepal_width_versicolor, desvio_padrao_sepal_width_versicolor,
 variancia_petal_length_versicolor, desvio_padrao_petal_length_versicolor,
 variancia_petal_width_versicolor, desvio_padrao_petal_width_versicolor) = calcular_variancia_desvio(versicolor_data_train_norm, media_sepal_length_versicolor, media_sepal_width_versicolor, media_petal_length_versicolor, media_petal_width_versicolor)

# Para Virginica
(variancia_sepal_length_virginica, desvio_padrao_sepal_length_virginica,
 variancia_sepal_width_virginica, desvio_padrao_sepal_width_virginica,
 variancia_petal_length_virginica, desvio_padrao_petal_length_virginica,
 variancia_petal_width_virginica, desvio_padrao_petal_width_virginica) = calcular_variancia_desvio(virginica_data_train_norm, media_sepal_length_virginica, media_sepal_width_virginica, media_petal_length_virginica, media_petal_width_virginica)

print("Setosa:")
print(f"Variância e Desvio Padrão para Sepal Length: {variancia_sepal_length_setosa}, {desvio_padrao_sepal_length_setosa}")
print(f"Variância e Desvio Padrão para Sepal Width: {variancia_sepal_width_setosa}, {desvio_padrao_sepal_width_setosa}")
print(f"Variância e Desvio Padrão para Petal Length: {variancia_petal_length_setosa}, {desvio_padrao_petal_length_setosa}")
print(f"Variância e Desvio Padrão para Petal Width: {variancia_petal_width_setosa}, {desvio_padrao_petal_width_setosa}")
print()

print("Versicolor:")
print(f"Variância e Desvio Padrão para Sepal Length: {variancia_sepal_length_versicolor}, {desvio_padrao_sepal_length_versicolor}")
print(f"Variância e Desvio Padrão para Sepal Width: {variancia_sepal_width_versicolor}, {desvio_padrao_sepal_width_versicolor}")
print(f"Variância e Desvio Padrão para Petal Length: {variancia_petal_length_versicolor}, {desvio_padrao_petal_length_versicolor}")
print(f"Variância e Desvio Padrão para Petal Width: {variancia_petal_width_versicolor}, {desvio_padrao_petal_width_versicolor}")
print()

print("Virginica:")
print(f"Variância e Desvio Padrão para Sepal Length: {variancia_sepal_length_virginica}, {desvio_padrao_sepal_length_virginica}")
print(f"Variância e Desvio Padrão para Sepal Width: {variancia_sepal_width_virginica}, {desvio_padrao_sepal_width_virginica}")
print(f"Variância e Desvio Padrão para Petal Length: {variancia_petal_length_virginica}, {desvio_padrao_petal_length_virginica}")
print(f"Variância e Desvio Padrão para Petal Width: {variancia_petal_width_virginica}, {desvio_padrao_petal_width_virginica}")
```

### Calcular Likelihood do Atributo e fazer Distribuição Gaussiana
 **Cálculo da probabilidade Gaussiana**:
   A fórmula da densidade de probabilidade para uma distribuição normal é:

   $$
   p(x|y) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2 \sigma^2}\right)
   $$

  Onde:
- μ é a média da característica.
- σ é o desvio padrão da característica.
- x é o valor da característica.


```python
import math

def calcular_likelihood_atributo(valor_atributo, media_atributo, desvio_padrao_atributo):
    return (1 / (math.sqrt(2 * math.pi * desvio_padrao_atributo**2))) * math.exp(-((valor_atributo - media_atributo)**2) / (2 * desvio_padrao_atributo**2))

# Calcular likelihood para Setosa
likelihood_sepal_length_setosa = [calcular_likelihood_atributo(length, media_sepal_length_setosa, desvio_padrao_sepal_length_setosa) for length in setosa_data_train_norm[:, 0]]
likelihood_sepal_width_setosa = [calcular_likelihood_atributo(width, media_sepal_width_setosa, desvio_padrao_sepal_width_setosa) for width in setosa_data_train_norm[:, 1]]
likelihood_petal_length_setosa = [calcular_likelihood_atributo(length, media_petal_length_setosa, desvio_padrao_petal_length_setosa) for length in setosa_data_train_norm[:, 2]]
likelihood_petal_width_setosa = [calcular_likelihood_atributo(width, media_petal_width_setosa, desvio_padrao_petal_width_setosa) for width in setosa_data_train_norm[:, 3]]

# Calcular likelihood para Versicolor
likelihood_sepal_length_versicolor = [calcular_likelihood_atributo(length, media_sepal_length_versicolor, desvio_padrao_sepal_length_versicolor) for length in versicolor_data_train_norm[:, 0]]
likelihood_sepal_width_versicolor = [calcular_likelihood_atributo(width, media_sepal_width_versicolor, desvio_padrao_sepal_width_versicolor) for width in versicolor_data_train_norm[:, 1]]
likelihood_petal_length_versicolor = [calcular_likelihood_atributo(length, media_petal_length_versicolor, desvio_padrao_petal_length_versicolor) for length in versicolor_data_train_norm[:, 2]]
likelihood_petal_width_versicolor = [calcular_likelihood_atributo(width, media_petal_width_versicolor, desvio_padrao_petal_width_versicolor) for width in versicolor_data_train_norm[:, 3]]

# Calcular likelihood para Virginica
likelihood_sepal_length_virginica = [calcular_likelihood_atributo(length, media_sepal_length_virginica, desvio_padrao_sepal_length_virginica) for length in virginica_data_train_norm[:, 0]]
likelihood_sepal_width_virginica = [calcular_likelihood_atributo(width, media_sepal_width_virginica, desvio_padrao_sepal_width_virginica) for width in virginica_data_train_norm[:, 1]]
likelihood_petal_length_virginica = [calcular_likelihood_atributo(length, media_petal_length_virginica, desvio_padrao_petal_length_virginica) for length in virginica_data_train_norm[:, 2]]
likelihood_petal_width_virginica = [calcular_likelihood_atributo(width, media_petal_width_virginica, desvio_padrao_petal_width_virginica) for width in virginica_data_train_norm[:, 3]]

# Exemplo de impressão para Setosa
print("Setosa:")
print(f"Likelihood Sepal Length: {likelihood_sepal_length_setosa}")
print(f"Likelihood Sepal Width: {likelihood_sepal_width_setosa}")
print(f"Likelihood Petal Length: {likelihood_petal_length_setosa}")
print(f"Likelihood Petal Width: {likelihood_petal_width_setosa}")
print()

# Exemplo de impressão para Versicolor
print("Versicolor:")
print(f"Likelihood Sepal Length: {likelihood_sepal_length_versicolor}")
print(f"Likelihood Sepal Width: {likelihood_sepal_width_versicolor}")
print(f"Likelihood Petal Length: {likelihood_petal_length_versicolor}")
print(f"Likelihood Petal Width: {likelihood_petal_width_versicolor}")
print()

# Exemplo de impressão para Virginica
print("Virginica:")
print(f"Likelihood Sepal Length: {likelihood_sepal_length_virginica}")
print(f"Likelihood Sepal Width: {likelihood_sepal_width_virginica}")
print(f"Likelihood Petal Length: {likelihood_petal_length_virginica}")
print(f"Likelihood Petal Width: {likelihood_petal_width_virginica}")
```

# Algoritmo Naive Bayes Gaussiano com teste

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix

# Carregar o dataset Iris
iris = datasets.load_iris()
data, target = iris.data, iris.target
iris_df = pd.DataFrame(data, columns=iris.feature_names)
iris_df['class'] = target
iris_df['class'] = iris_df['class'].map(dict(enumerate(iris.target_names)))

# Funções para calcular as médias e desvios padrão
def calcular_media_atributos(data):
    return np.mean(data, axis=0)

def calcular_variancia_desvio(data, medias):
    variancias = np.var(data, axis=0)
    desvios = np.sqrt(variancias)
    return variancias, desvios

def calcular_likelihood_atributo(valor_atributo, media_atributo, desvio_padrao_atributo):
    return (1 / (np.sqrt(2 * np.pi * desvio_padrao_atributo**2))) * np.exp(-((valor_atributo - media_atributo)**2) / (2 * desvio_padrao_atributo**2))

# Função para calcular as probabilidades a priori
def calcular_prior_probabilidades(classes):
    unique, counts = np.unique(classes, return_counts=True)
    total = len(classes)
    return dict(zip(unique, counts / total))

# Função para prever a classe de uma amostra
def prever_classe(sample, medias, desvios, priors):
    likelihoods = {}
    for classe in medias.keys():
        likelihood = np.prod([calcular_likelihood_atributo(sample[i], medias[classe][i], desvios[classe][i]) for i in range(len(sample))])
        likelihoods[classe] = likelihood * priors[classe]
    return max(likelihoods, key=likelihoods.get)

# Dividir os dados em treino e teste
train_data, test_data, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)

# Normalizar os dados
scaler = StandardScaler()
train_data_norm = scaler.fit_transform(train_data)
test_data_norm = scaler.transform(test_data)

# Calcular as médias, variâncias e desvios padrão para cada classe
medias = {}
desvios = {}
for classe in np.unique(train_target):
    classe_data = train_data_norm[train_target == classe]
    medias[classe] = calcular_media_atributos(classe_data)
    variancias, desvios[classe] = calcular_variancia_desvio(classe_data, medias[classe])

# Calcular as probabilidades a priori
priors = calcular_prior_probabilidades(train_target)

# Prever as classes para os dados de teste
previsoes = [prever_classe(sample, medias, desvios, priors) for sample in test_data_norm]

# Visualização das distribuições de probabilidade
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

atributos = iris.feature_names
classes = iris.target_names

for i, atributo in enumerate(atributos):
    row, col = divmod(i, 2)
    for classe in classes:
        sns.kdeplot(train_data_norm[train_target == list(classes).index(classe), i], label=classe, ax=axs[row, col])
    axs[row, col].set_title(f'Distribuição do {atributo}')
    axs[row, col].legend()

plt.show()

# Visualização da classificação
plt.figure(figsize=(10, 6))

# Usando Petal Length e Petal Width para visualização
plt.scatter(test_data[:, 2], test_data[:, 3], c=test_target, cmap='viridis', marker='o', label='Verdadeiro')
plt.scatter(test_data[:, 2], test_data[:, 3], c=previsoes, cmap='cool', marker='x', label='Previsto')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Classificação Verdadeira vs Prevista')
plt.legend()
plt.show()
# Calcular a matriz de confusão
cm = confusion_matrix(test_target, previsoes)

# Contagem de acertos e erros
acertos = np.diag(cm)
erros = np.sum(cm, axis=1) - acertos

# Labels das classes
labels = iris.target_names

# Plotar gráfico de barras de acertos e erros
fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.35
x = np.arange(len(labels))

# Plotagem das barras
rects1 = ax.bar(x - bar_width/2, acertos, bar_width, label='Acertos')
rects2 = ax.bar(x + bar_width/2, erros, bar_width, label='Erros')

# Adicionar rótulos, títulos e legenda
ax.set_xlabel('Classe')
ax.set_ylabel('Quantidade')
ax.set_title('Acertos e Erros na Classificação do Dataset Iris')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Adicionar textos com quantidades
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.0f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# Exibir gráfico
fig.tight_layout()
plt.show()
```
